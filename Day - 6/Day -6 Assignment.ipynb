# Question 1 : 
    For this challenge,create a bank account class that has two attributes
    *ownerName
    *Balance
    And two methods
    *deposit
    *withdraw
    As an added requirement,withdrawals may not exceed the available balance.
    Instantiate your class,make several deposits and withdrawals,and test to
    make sure the account cant be overdrawn.

```python
class bankAccount():
    def __init__(self, ownerName, balance):
        self.ownerName = ownerName
        self.balance = balance
        
    def deposit(self):
        addMoney = float(input("Enter the amount you want to Deposit : "))
        self.balance += addMoney
        print("\nYou have successfully Deposited the money : ", addMoney)
  
    def withdraw(self):
        
        withdrawMoney = float(input("\nEnter the amount you want to withdraw : "))
            
        if self.balance >= withdrawMoney: 
            self.balance -= withdrawMoney 
            print("\nYou have Successfully Withdrawn the money : ", withdrawMoney) 
        else: 
            print("\nInsufficient Balance in the Account...") 

myBank = bankAccount("Abhishek Raj", 500000)

print(" -----My Bank Account Details ----- : ")
print("Account Holders Name : ", myBank.ownerName)
print("Initial Balance in the Account : ", myBank.balance)

myBank.deposit()
print("\n")
myBank.withdraw()




print("\nThe Net Amount in the Bank is : ", myBank.balance)

Continue = "y"
while Continue.lower() == "y":
    myBank.deposit()
    myBank.withdraw()
    print("\nThe Net Amount in the Bank is : ", myBank.balance)
    Continue = input("\nYou Want to make another Deposite/Withdwraw Y/N : ")

       
```

# Question 2 :
    For this challenge,create a cone class that has two attributes:
    *R=Radius
    *h=Height
    And two methods:
    *Volume = Π * r2 = (h/3)
    *Surface area : base : Π * r2 , side : Π * r * √(r2 + h2)
    Make only one class with functions,as in where required import Math.

```python
import math


#l**2 = h**2 + r**2

pi = math.pi
class cone():
    def __init__(self, r, h):
        self.r = r
        self.h = h
    
    def coneVolume(self):
        vol = (1/3) * self.h * pi * self.r**2
        print("\nVolume of the Cone whose Radius is :", {self.r}, "and Height is :", {self.h}, "is : ", vol)
    
    def coneSurfaceArea(self):
        
        SA = pi * self.r * (self.r + math.sqrt(self.r**2 + self.h**2))
        print("\nSurface Area of the Cone whose Radius is :", {self.r}, "and Height is :", {self.h}, "is : ", SA)


radius = float(input("\nEnter the Radius of cone : "))

height = float(input("\nEnter the Height of cone : "))

c = cone(radius, height)

c.coneVolume()

c.coneSurfaceArea()
```
